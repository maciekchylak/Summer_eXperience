## Różnica w czasie realizacji zamówienia a estymowalnym czasem

```python

data_merged[['order_delivered_customer_date','order_delivered_customer_sec']] = data_merged["order_delivered_customer_date"].str.split(' ', 1, expand=True)
data_merged[['order_estimated_delivery_date',"order_estimated_delivery_sec"]] = data_merged["order_estimated_delivery_date"].str.split(' ', 1, expand=True)
data_merged = data_merged.drop(["customer_unique_id","customer_id", "order_id", "seller_id","review_id", "order_purchase_timestamp", "order_delivered_customer_sec", "order_estimated_delivery_sec"], axis=1)
```

```python
import datetime
data_merged["order_delivered_customer_date"] = pd.to_datetime(data_merged.order_delivered_customer_date)
data_merged["order_estimated_delivery_date"] = pd.to_datetime(data_merged.order_estimated_delivery_date)
```

```python
data_merged["delivery_delta"] = (data_merged["order_estimated_delivery_date"] - data_merged["order_delivered_customer_date"])
data_merged["delivery_delta"] = pd.to_numeric(data_merged["delivery_delta"].dt.days, downcast='integer')
```

```python
data_merged = data_merged.drop(["order_delivered_customer_date", "order_estimated_delivery_date"], axis=1)
```

```python
data_merged.head()
```

Dla każdego miasta, regionu oraz kategorii stwórzmy w pewnym sensie ocene produktu. Taką,aby była ona jak w najmniejszym stopniu powiązana z różnicą w realizacji zamówienia oraz procentowej cenie przesyłki w odniesieniu do ceny produktu 

W przypadku ostatecznej oceny jakości nie będą nas interesowały zamówienia, w których czas oczekiwania był bardzo wydłużony, a ostateczna ocena zamówienia niska, gdyż istnieje ogromne prawdopodobieństwo, że właśnie wydłużony czas realizacji był głównym powodem, dla którego zdecydowano się wystawic tak niską ocenę.

Również dotyczy to przypadku, odwrotnym - bardzo skrócony czas oczekiwania i wysoka ocena

```python
delivery_time_minus = data_merged[(data_merged.delivery_delta < 0)]
delivery_time_minus  = delivery_time_minus["delivery_delta"] 
```

```python
delivery_time_minus = delivery_time_minus.quantile(q=0.1, interpolation='linear')
delivery_time_minus
```

```python
delivery_time_plus = data_merged[(data_merged.delivery_delta > 0)]
delivery_time_plus  = delivery_time_plus["delivery_delta"] 
```

```python
delivery_time_plus = delivery_time_plus.quantile(q=0.9, interpolation='linear')
delivery_time_plus
```

Warto zauważyć, że w przypadku zamówień w Brazylii generalnie czasy oczekiwania na zamówienie są o wiele wyższe niż w krajach europejskich. Jako szczególnie wydłużone zamówienia przyjąłem więc górne 10%

```python
data_merged.drop(data_merged[(data_merged.delivery_delta >= delivery_time_plus) & (data_merged.review_score > 3)].index, inplace=True)
data_merged.drop(data_merged[(data_merged.delivery_delta <= delivery_time_minus) & (data_merged.review_score > 3)].index, inplace=True)
```

```python
data_merged["freigth/price"] = ((data_merged["freigth/price"] * 5) + 0.4)
```

Stwórzmy także wcześniej wspomnianą ogólną skale oceny. Będzie nią iloraz oceny zamowienia przez stosunek ceny przesyłki do ceny produktu. Aby to zrobić musiałem stransormować zmienną freigth/price, aby nie występowały w niej wartości zerowe

```python

data_merged["final_score"] = (data_merged["review_score"] / data_merged["freigth/price"])
```

```python
data_merged = data_merged.drop(["review_score", "freigth/price", "delivery_delta"], axis=1)
```

```python
home =['air_conditioning','garden_tools', 'kitchen_dining_laundry_garden_furniture',  'home_construction', 'home_appliances_2', 'costruction_tools_tools',  'la_cuisine', 'construction_tools_lights', 'costruction_tools_garden', 'small_appliances_home_oven_and_coffee', 'construction_tools_construction']
work_and_study =['office_furniture', 'stationery', 'books_technical', 'security_and_services', 'agro_industry_and_commerce', 'construction_tools_safety', 'market_place',  'signaling_and_security',  'industry_commerce_and_business', 'housewares']
decoration = ['furniture_decor','furniture_living_room', 'furniture_bedroom', 'home_comfort_2', 'home_confort','flowers', 'furniture_mattress_and_upholstery', 'bed_bath_table', 'arts_and_craftmanship']
beauty = ['health_beauty','fashio_female_clothing', 'perfumery', 'fashion_male_clothing', 'diapers_and_hygiene', 'fashion_childrens_clothes',  'fashion_underwear_beach', 'fashion_bags_accessories', 'fashion_shoes', 'luggage_accessories']
devices = ['computers_accessories', 'telephony','watches_gifts', 'fixed_telephony', 'pc_gamer', 'small_appliances', 'portable_kitchen_and_food_preparers', 'computers', 'electronics', 'tablets_printing_image', 'home_appliances']
free_time = ['sports_leisure', 'fashion_sport', 'art', 'toys', 'consoles_games', 'books_general_interest', 'musical_instruments', 'dvds_blu_ray', 'music', 'cine_photo',  'cds_dvds_musicals',  'audio', 'books_imported']
others = ['others', 'auto', 'baby','food','pet_shop' , 'drinks','food_drink', 'cool_stuff'  ,'party_supplies', 'christmas_supplies']
all = home + work_and_study + decoration + beauty + devices + free_time + others
```

```python
dicts = [home, work_and_study, decoration, beauty, devices, free_time, others]
dicts_names = ["home", "work_and_study", "decoration", "beauty", "devices", "free_time", "others"]
def dict_update(a, x, i):
    dicts[i] = dict(zip(x, len(x)*[a]))
```

```python
for i in range(len(dicts_names)):
    dict_update(dicts_names[i], dicts[i], i)
```

```python
sub_category_dict = dicts[0]
for i in range(1,len(dicts)):
   sub_category_dict  = {**sub_category_dict, **dicts[i]} 
```

```python
inv_sub_category_dict = {"home":  home, "work_and_study": work_and_study,  "decoration": decoration,"beauty": beauty, "devices": devices, "free_time": free_time, "others": others}
```

## Rozwiązanie 

Ostateczny pomysł na rozwiązanie problemu w przypadku podania informacji:

* Wybranie po dwa najchętniej zakupowane produkty z podanych przez użytkownika kategorii
* Wybranie jednego produktu najchętniej kupowanego w jego regionie
* Wybranie po jednym produkcie z kategorii podobnych do tych, które podał w preferowanych kategoriach 

W przypadku nie podania informacji:
* Wybranie po jednym najchętniej sprzedających się produktów z różnych kategorii

```python
import random
def result(cust_pref, cust_region):
    result = []
    tmp = []
    if len(cust_pref) > 0 and len(cust_region) > 0:
        for cat in cust_pref:
            cat_trans = max(prod_cat_en.loc[prod_cat_en["product_category_name"] == cat,"product_category_name_english"])
            tmp = best_in_category.loc[best_in_category["product_category_name_english"] == cat_trans,"product_id"]
            for el in tmp:
                result.append(el)
            huge_cat = sub_category_dict[cat_trans]
            huge_cat_list = inv_sub_category_dict[huge_cat]
            name_cat = cat_trans
            name_prod = result[0]
            while (name_cat in cust_pref) or (name_prod in result):
                name_cat = random.choice(huge_cat_list)
                name_prod = max(best_in_category.loc[best_in_category["product_category_name_english"] == name_cat,"product_id"])
                
            result.append(name_prod)

        tmp = best_in_region.loc[best_in_region["customer_state"] == cust_region[1],"product_id"]
        for el in tmp:
            result.append(el)
    else:
        category = ["abc"]
        name = "sports_leisure"
        tmp_dict = {"home": 0,"work_and_study": 0, "decoration" : 0,"beauty": 0, "devices":0,"free_time": 0, "others": 0}
        for i in range(10):
            while (tmp_dict[sub_category_dict[name]] > 2) or (name in category):
                name = random.choice(all)                
            tmp_dict[sub_category_dict[name]] += 1
            category.append(name)
            result.append(max(best_in_category.loc[best_in_category["product_category_name_english"] == name,"product_id"]))
        
    return result
        
        
        
    
```
